```python
import math
for i in range(1, 9, 1):

    true_values = [row[0] for row in bond_lengths]
    measured_values = [row[i] for row in bond_lengths]
    deviations = abs(np.array(measured_values) - np.array(true_values))
    
    plt.figure(figsize=(10, 3))
    plt.plot(true_values, label='MWS', marker='o')
    plt.plot(measured_values, label= f'{theory[i]}', marker='x')
    plt.title('Bond length comparison')
    plt.ylabel('Bond length, Å')
    
    plt.legend()
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(10, 3))
    plt.plot(deviations, label='deviations', marker='s')
    
    plt.title('deviations')
    plt.ylabel('Bond length, Å')
    
    plt.grid()
    plt.show()

    EV = [
    true_values
    ] #Experimental values
    CV = [
    measured_values
    ] #Calculated values

    MSE = np.square(np.subtract(EV,CV)).mean() 
 
    RMSE = math.sqrt(MSE)
    print(f'Root Mean Square Error {theory[i]}, Å:\n')
    print(RMSE)
```
